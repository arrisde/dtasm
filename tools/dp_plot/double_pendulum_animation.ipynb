```python
from io import StringIO
import csv

from numpy import sin, cos
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# either load csv from a file:
#with open('../../runtime/examples/dtasmtime_rs/out.csv') as csv_file:
# or from a string literal:
with StringIO(result_csv) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    angle_list = []
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            angle_list.append((float(row[0]), float(row[1]), float(row[3])))
            line_count += 1
    print(f'Read {line_count} lines.')

y = np.array(angle_list)

L1 = 1.0  # length of pendulum 1 in m
L2 = 1.0  # length of pendulum 2 in m
dt = y[1,0] - y[0,0]

x1 = L1*sin(y[:, 1])
y1 = -L1*cos(y[:, 1])

x2 = L2*sin(y[:, 2]) + x1
y2 = -L2*cos(y[:, 2]) + y1

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, autoscale_on=False, xlim=(-2, 2), ylim=(-2, 1))
ax.set_aspect('equal')
ax.grid()

line, = ax.plot([], [], 'o-', lw=2)
time_template = 'time = %.1fs'
time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes)

def animate(i):
    thisx = [0, x1[i], x2[i]]
    thisy = [0, y1[i], y2[i]]

    line.set_data(thisx, thisy)
    time_text.set_text(time_template % (i*dt))
    return line, time_text

ani = animation.FuncAnimation(
    fig, animate, len(y), interval=dt*1000, blit=True, repeat=False)
plt.show()
```
